# Step 3 : Data Cleaning

```python
# convert timestamp to standard datetime format

from datetime import datetime
merged_data['timestamp'] = pd.to_datetime(merged_data['timestamp'], errors='coerce', format='%d/%m/%Y %H:%M')
```

```python
import pandas as pd
import os
import pickle
from sklearn.preprocessing import MinMaxScaler

# Columns to scale
scaling_columns = [
    "disk_io", "network_latency", "node_temperature", "node_cpu_usage", 
    "node_memory_usage", "cpu_request", "cpu_limit", "memory_request", 
    "memory_limit", "cpu_usage", "memory_usage", "network_bandwidth_usage"
]

# Directory to save scaler models
scaler_directory = '/Users/zxgan/FYP_Kubernetes/Scalers/'
os.makedirs(scaler_directory, exist_ok=True)

# Create and save scalers for each column
scalers = {}

for column in scaling_columns:
    scaler = MinMaxScaler()
    merged_data[[column]] = scaler.fit_transform(merged_data[[column]])
    scalers[column] = scaler
    # Save the scaler
    scaler_path = os.path.join(scaler_directory, f"{column}_scaler.pkl")
    with open(scaler_path, 'wb') as f:
        pickle.dump(scaler, f)

print("Scaling complete and scalers saved.")

# Show scaled data
print(merged_data.head(20))
```

```python
# perform one-hot encoding on the dataset
categories = {
    "pod_status": ["Running", "Pending", "Succeeded", "Failed", "Unknown"],
}

# One-hot encode columns with pre-specified categories
for column, unique_values in categories.items():
    for value in unique_values:
        col_name = f"{column}_{value}"
        if col_name not in merged_data.columns:
            merged_data[col_name] = (merged_data[column] == value).astype(int) if column in merged_data else 0


merged_data = merged_data.drop(columns=categories.keys(), errors='ignore')

merged_data.head(20)
```

```python
#drop the pod_name, namespace_x, namespace_y, event_type, event_message, scalling_event, pod_lifetime_seconds, restart_count, uptime_seconds, deployment_strategy, scaling_policy, 

cleaned_data = merged_data.drop(columns=['pod_name', 'namespace_x', 'namespace_y', 'event_type', 'event_message', 'scaling_event', 'pod_lifetime_seconds', 'restart_count', 'uptime_seconds', 'deployment_strategy', 'scaling_policy'], errors='ignore')

cleaned_data.head(20)
```

```python
#export the cleaned dataset

cleaned_data.to_csv('/Users/zxgan/FYP_Kubernetes/Dataset/cleaned_kubernetes_dataset.csv', index=False)
```

```python
#divide the dataset by node_name

node_names = cleaned_data["node_name"].unique()
node_data = {}
for node_name in node_names:
    node_data[node_name] = cleaned_data[cleaned_data["node_name"] == node_name]

#export the divided dataset
for node_name, data in node_data.items():
    data.to_csv(f'/Users/zxgan/FYP_Kubernetes/Dataset/cleaned_dataset/node_{node_name}_dataset.csv', index=False)
```
