# Step 1 : Load Data

```python
import pandas as pd

performance_metrics_path = '/Users/zxgan/FYP_Kubernetes/Dataset/kubernetes_performance_metrics_dataset.csv'
resource_allocation_path = '/Users/zxgan/FYP_Kubernetes/Dataset/kubernetes_resource_allocation_dataset.csv'

performance_metrics = pd.read_csv(performance_metrics_path)
resource_allocation = pd.read_csv(resource_allocation_path)

merged_data = pd.merge(performance_metrics, resource_allocation, on=['pod_name'], how='inner')

merged_data.to_csv('/Users/zxgan/FYP_Kubernetes/Dataset/merged_kubernetes_dataset.csv', index=False)
```

# Step 3 : Data Cleaning

```python
# convert timestamp to standard datetime format

from datetime import datetime
merged_data['timestamp'] = pd.to_datetime(merged_data['timestamp'], errors='coerce', format='%d/%m/%Y %H:%M')
```

```python
#drop the pod_name, namespace_x, namespace_y, event_type, event_message, scalling_event, pod_lifetime_seconds, restart_count, uptime_seconds, deployment_strategy, scaling_policy, 

cleaned_data = merged_data.drop(columns=['pod_name', 'namespace_x', 'namespace_y', 'event_type', 'event_message', 'scaling_event', 'pod_lifetime_seconds', 'restart_count', 'uptime_seconds', 'deployment_strategy', 'scaling_policy','pod_status'], errors='ignore')

cleaned_data.head(20)
```

```python
#divide the dataset by node_name

node_names = cleaned_data["node_name"].unique()
node_data = {}
for node_name in node_names:
    node_data[node_name] = cleaned_data[cleaned_data["node_name"] == node_name]

#export the divided dataset
for node_name, data in node_data.items():
    data.to_csv(f'/Users/zxgan/FYP_Kubernetes/Dataset/feature_engineer_dataset/node_{node_name}_dataset.csv', index=False)
```
